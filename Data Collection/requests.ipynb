```python
import requests
```

```python
for i in range(1,51):
    ht=requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open(f"htmls/page{i}.html","w",encoding="utf-8") as f:
        f.write(ht.text)
        if(ht.status_code==200):
            print("HTML fetched successfully")
        else:
            print("Some error occured")
```
