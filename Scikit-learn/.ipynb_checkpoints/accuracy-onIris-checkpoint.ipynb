```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data=pd.read_excel("Iris-train.xlsx")
data
```

```python
x_train=data.iloc[:,:-1]
y_train=data.iloc[:,-1]
```

```python
x_train
```

```python
y_train
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
prediction1=model.predict([[5.4,3,4.5,1.5]])
```

```python
prediction1
```

```python
datatest=pd.read_excel("Iris-test.xlsx")
x_test=datatest.iloc[:,:-1]

prediction2=model.predict(x_test.values)
```

```python
prediction2
```
