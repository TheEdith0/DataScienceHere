```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
#here we use list of lists
#Here we consider 0=red and orange=1 & apple weight more than oranges ->assumptions
features=[[150,0],
         [180,0],
         [120,1],
         [140,1]]
labels =["apple","apple","orange","orange"]

clf=DecisionTreeClassifier()
clf.fit(features,labels)
prediction = clf.predict([[160,0]])

print(f"The fruit is {prediction[0]}")
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

plot_tree(clf,feature_names=["Weight","Color"],class_names=["apple","orange"])
```
