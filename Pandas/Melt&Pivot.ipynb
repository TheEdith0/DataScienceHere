```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
df2=df.melt(id_vars=["Name"],value_vars=["Math","Science","English"],var_name="Subject",value_name="Marks")
print(df2)
```

```python
df3=df2.pivot(index="Name",columns="Subject",values="Marks")# cannot handle ssame row dtat with different value columns
print(df3)
```

```python
df2.loc[9]={"Name":"Alice","Subject":"Math","Marks":90}
df2
```

```python
df4=df2.pivot_table(index="Name",columns="Subject",values="Marks",aggfunc="mean")
df4
```
