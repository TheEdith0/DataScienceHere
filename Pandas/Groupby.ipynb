```python
import pandas as pd
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  
print(df)
```

```python
df.groupby("Department")["Salary"].sum()# this return series
```

```python
df.groupby("Team")["Salary"].mean()
```

```python
df.groupby("Team")["Salary"].max()
```

```python
df.groupby("Department")["Salary"].min() #["Salary"].min() -> list or dictionary
```

```python
df.groupby("Team")["Salary"].agg(["max","min","mean"]) # .aggregate() also same 
```

```python
df.groupby("Team")["Salary"].agg(
    Average="mean",
    Maximumm="max"
)
```

```python
df.groupby("Department").agg({
    "Salary":"mean",
    "Age":"max"
})
```

```python
df["Team Avg"] = df.groupby("Team")["Salary"].transform("mean")
df
```

```python
df.groupby("Department").filter(lambda x: x["Salary"].mean() >85)
```
