```python
import numpy as np
arr1=np.array([[1,2,3],[8,9,7],[12,19,4]])
print("shape",arr1.shape)
```

```python
arrINT=np.array([1.1,2.2,3.3])
print(arrINT.dtype)
```

```python
arr2=np.arange(1,7,1)
print(arr2)
print(arr2[[1,3,5]])#fancy indexing 
```

```python
arr3=np.ones((1,9))
print(arr3)
arr4=arr3.reshape((3,3))
print(arr4)
```
