```python
import numpy as np
import time

size = 1_000_00
list1=list(range(size))
list2=list(range(size))

start=time.time()
result=[x+y for x,y in zip(list1,list2)]
end= time.time()
print(end-start)

arr1=np.array(list1)
arr2=np.array(list2)

start=time.time()
arr3=arr1+arr2 #vectorization
end=time.time()
print(end-start)

#When you take data type of less bits it took less time to process
arr4=np.array(list1,dtype=np.int32)
arr5=np.array(list2,dtype=np.int32)

start=time.time()
arr6=arr4+arr5 #vectorization
end=time.time()
print(end-start)
```
