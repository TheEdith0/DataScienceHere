```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm

x=np.linspace(-10,10,1000)

mean=0
std=2

y = norm.pdf(x,loc=mean,scale=std)

plt.plot(x,y)
plt.title("Standard Normal Distribution (μ=0, σ=2)")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()

```
