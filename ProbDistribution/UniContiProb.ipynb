```python
#plot a histogram for PDF in picking a no. between 0 to 1
import numpy as np 
import matplotlib.pyplot as plt
```

```python
sample=np.random.uniform(0,1,1000000)

plt.hist(sample,bins=50,alpha=0.6,color="skyblue",density=True)
plt.title("Continuous Uniform distribution from 0 to 1")
plt.xlabel("Range of no. picked")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
# the way we move to infinte no. are picked we will get a flat line
```
