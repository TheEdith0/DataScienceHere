```python
import json 
def load_file(filename):
    with open(filename,"r") as file:
        data = json.load(file)
    return data
```

```python
def clean_file(raw_data):
    raw_data["users"]=[user for user in raw_data["users"] if user["name"].strip()]
    
    for user in raw_data["users"]:
        user["friends"]=list(set(user["friends"]))
        
    raw_data["users"]=[user for user in raw_data["users"] if user["friends"] or user["liked_pages"]]
    
    unique_dict={}
    for page in raw_data["pages"]:
        unique_dict[page["id"]]=page
    raw_data["pages"]=list(unique_dict.values())    
    
    with open("Cleaned_data.json","w") as file:
        json.dump(raw_data,file,indent = 4)
    
    print("Data Clenaed Successfully!!")   
data=load_file("Data2.json")
clean_file(data)
```
