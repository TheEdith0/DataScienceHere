```python
import json 
def load_data(filename):
    with open(filename,"r") as file:
        data=json.load(file)
    return data

def people_maybein_contact(user_id,data):
    user_friends={}
    for user in data["users"]:
        user_friends[user["id"]]=set(user["friends"])
    if user_id not in user_friends.keys():
        return []
        print("User do not exist!!!")
    direct_friends=user_friends[user_id]
    suggestions={}
    for direct_friend in direct_friends:
        for friend in user_friends[direct_friend]:
            if (friend != user_id and friend not in direct_friends ):
                suggestions[friend]=suggestions.get(friend,0) + 1
    sorted_suggestions=dict(sorted(suggestions.items(),key= lambda x:x[1],reverse=True))#return list of tuples
    return [rec for rec in sorted_suggestions.keys()]
    
                        
data=load_data("Data.json")
user_id=int(input("Enter the user_id: "))
recommend=people_maybein_contact(user_id,data)
print(f"Friends of User: {user_id} are {recommend}")
```
