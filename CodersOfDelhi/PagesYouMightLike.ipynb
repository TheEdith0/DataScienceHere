```python
import json 
def load_data(filename):
    with open(filename,"r") as file:
        data=json.load(file)
    return data

def recommend_pages(user_id,data):
    page_dict={}
    for user in data["users"]:
        page_dict[user["id"]]=user["liked_pages"]
    if user_id not in page_dict.keys():
        return []
    direct_liked=page_dict[user_id]
    suggestions={}
    for pages in page_dict.values():
        if (len(list(set(direct_liked) & set(pages))) !=0) :
            suggestions[tuple((set(direct_liked) | set(pages))-(set(direct_liked) & set(pages)))] =(len(list(set(direct_liked) & set(pages))))
    sorted_suggestions=sorted(suggestions.items(),key=lambda x:x[1],reverse=True)
    list_to_suggest=[]
    for sug,_ in sorted_suggestions:
        list_to_suggest = list(set(list_to_suggest) | set(sug))
    final=list(set(list_to_suggest) -set(direct_liked))
    return final
                
data=load_data("massive_data.json")
user_id=int(input("Enter your user ID: "))
rec=recommend_pages(user_id,data)
print(f"User no :{user_id} might like pages are {rec}")
```
