```python
import json

# Load JSON data from file
def load_data(filename):
    with open(filename, "r") as file:
        return json.load(file)

# Get page name from ID
def get_page_name(page_id, data):
    for page in data["pages"]:
        if page["id"] == page_id:
            return page["name"]
    return "Unknown Page"

# Recommend pages based on common interests
def find_pages_you_might_like(user_id, data):
    # Build dictionary of user_id → set of liked pages
    user_pages = {user["id"]: set(user["liked_pages"]) for user in data["users"]}

    # Return empty list if user not found
    if user_id not in user_pages:
        return []

    user_liked_pages = user_pages[user_id]
    page_suggestions = {}

    # Compare with all other users
    for other_user_id, other_liked_pages in user_pages.items():
        if other_user_id == user_id:
            continue

        # Only consider users with at least one shared page
        shared_pages = user_liked_pages.intersection(other_liked_pages)
        if not shared_pages:
            continue

        # Suggest pages this user hasn't liked yet
        for page in other_liked_pages:
            if page not in user_liked_pages:
                page_suggestions[page] = page_suggestions.get(page, 0) + len(shared_pages)

    # Sort pages by score (higher = more shared interests)
    sorted_pages = sorted(page_suggestions.items(), key=lambda x: x[1], reverse=True)

    # Return only the page IDs (or change to names if preferred)
    return [page_id for page_id, _ in sorted_pages]

# ---------- RUNNING THE PROGRAM ----------

data = load_data("massive_data.json")
user_id = int(input("Enter your user ID: "))
page_recommendations = find_pages_you_might_like(user_id, data)

# Print page names, not just IDs
print(f"\n Pages You Might Like for User {user_id}:")

if not page_recommendations:
    print("No personalized recommendations found.")
else:
    for pid in page_recommendations:
        print(f"• {get_page_name(pid, data)} (ID: {pid})")
```
