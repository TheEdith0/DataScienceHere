```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

```python
tips=sns.load_dataset("tips")
```

```python
gap=px.data.gapminder()
```

```python
gap
```

```python
tips
```

```python
#scatterplot
sns.stripplot(data=tips,x="total_bill",y="tip",jitter=False)
# seaborn add noise automatically to it and to remove noise jitter=False
```

```python
sns.catplot(data=tips,x="total_bill",y="tip",kind="strip",hue="sex")
```

```python
#SWARMPLOT : almost same a strip plot
sns.swarmplot(data=tips,x="total_bill",y="tip")
```

```python
sns.catplot(data=tips,x="total_bill",y="tip",kind="swarm",hue="sex")
```

```python
#BOXPLOT 
sns.boxplot(data=tips,x="day",y="total_bill")
```

```python
sns.catplot(data=tips,x="day",y="total_bill",hue="smoker",kind="box")
```

```python
#SingleBoxplot
sns.boxplot(data=tips,y="total_bill")
```

```python
#VIOLINPLOT
sns.violinplot(data=tips,x="day",y="total_bill",hue="sex")
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="violin",hue="sex",split=True)
```

```python
#BARPLOT
import numpy as np
sns.barplot(data=tips,x="day",y="total_bill",hue="sex",estimator=np.median)   #ci=None remove error bars
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="bar",ci=None) # or you can use errorbar=None
```

```python
#POINTPLOT : connects the error bar of bar plot and remove bars 
sns.pointplot(data=tips,x="day",y="total_bill")
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="point")
```

```python
#COUNTPLOT:histogram with categorial(eg.- no. of female,no. of male)
sns.countplot(data=tips,x="sex",hue="smoker")
```

```python
sns.catplot(data=gap,x="continent",kind="count")
```
