```python
import seaborn as sns
import pandas as pd
import plotly.express as px #dataset library
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
gap=px.data.gapminder()
```

```python
#histogram
sns.histplot(data=tips,x="total_bill",hue="sex",)
```

```python
sns.displot(data=tips,x="total_bill",kind="hist")
```

```python
#kde plot
sns.kdeplot(data=tips,x="total_bill",hue="sex",fill=True)
```

```python
sns.displot(data=tips,x="total_bill",kind="kde")
```

```python
#rugplot
sns.rugplot(data=tips,x="total_bill",height=0.2,color="orange")
```

```python
sns.displot(data=tips,x="total_bill",kind="kde",rug=True) #used together mainly
```

```python
#bivariate histogram
sns.histplot(data=tips,x="total_bill",y="tip")
```

```python
#bivariate kde
sns.kdeplot(data=tips,x="total_bill",y="tip")
```
