```python
import seaborn as sns
import pandas as pd
import plotly.express as px #dataset library
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
#Multiplot : are just create using facet like we did in relational plots now here we are going to use other method.
#using FacetGrid
g=sns.FacetGrid(data=tips,col="day",row="time")
g.map(sns.violinplot,"sex","total_bill")
g.add_legend()
```

```python
#pairplot
iris=sns.load_dataset("iris")
sns.pairplot(iris)
# for same col and row histogram
```

```python
#PairGrid : more flexible than multi plot 

#no need of specifing row and col , it pairs all column and row accordingly(numeric)
#all scatterplots 
g1=sns.PairGrid(data=tips)
g1.map(sns.scatterplot)
```

```python
# diagnol and offdiagnol different
g2=sns.PairGrid(data=tips)
g2.map_diag(sns.histplot)
g2.map_offdiag(sns.lineplot)
```

```python
# upper, lower and diagnol different
#vars=[] : to control cars allowed
g3=sns.PairGrid(data=iris ,vars=["sepal_length","petal_length"])
g3.map_diag(sns.histplot)
g3.map_upper(sns.lineplot)
g3.map_lower(sns.barplot)
```

```python
#JOINTPLOT: hist + scatter used to plot 2 plots in one figure 
sns.jointplot(data=tips,x="total_bill",y="tip")
```

```python
# JointGrid: for your choice plots

g1=sns.JointGrid(data=tips,x="total_bill",y="tip")
g1.plot(sns.lineplot,sns.kdeplot)
```
