```python
import seaborn as sns
import matplotlib.pyplot as plt
print(sns.get_dataset_names())
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.lineplot(x="total_bill",y="tip",data=tips)
plt.title("tips")
plt.xlabel("Total Bills")
plt.ylabel("Tip")
plt.show()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("tips")
plt.xlabel("Total Bills")
plt.ylabel("Tip")
plt.show()
```

```python
sns.barplot(x="total_bill",y="tip",data=tips)
plt.title("tips")
plt.xlabel("Total Bills")
plt.ylabel("Tip")
plt.show()
```

```python
sns.boxplot(x="total_bill",y="tip",data=tips)
plt.title("tips")
plt.xlabel("Total Bills")
plt.ylabel("Tip")
plt.show()
```

```python
sns.histplot(x="total_bill",y="tip",data=tips)
plt.title("tips")
plt.xlabel("Total Bills")
plt.ylabel("Tip")
plt.show()
```

```python
flights=sns.load_dataset("flights")
flights
```

```python
p_table=flights.pivot(index="year",columns="month",values="passengers")
sns.heatmap(p_table,annot=True,fmt="d",cmap="YlGnBu")# fmt="d" means integer format
```

```python
import pandas as pd

df = pd.DataFrame({
    "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})

sns.scatterplot(x="age", y="salary", hue="gender", data=df)
plt.title('Salary vs Age Scatter Plot')
plt.show()
```
