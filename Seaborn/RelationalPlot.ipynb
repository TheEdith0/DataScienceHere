```python
import seaborn as sns
import pandas as pd
import plotly.express as px #dataset library
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
gap=px.data.gapminder()
gap
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="time",size="size") # axes level
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="sex",style="time",size="size",kind="scatter") # figure level
```

```python
sns.lineplot(data=tips,x="total_bill",y="tip",hue="sex",style="time",size="size") # axes level
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="sex",style="time",size="size",kind="line") #figure level
```

```python
# Facet plot: mutiple plot in one figure using coland row/col_wrap
gap2 = gap[gap["country"].isin(["India", "China", "Egypt"])]
sns.relplot(data=gap2,x="year",y="lifeExp",col="country",row="continent",kind="line")
```
