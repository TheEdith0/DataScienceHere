```python
#Virat Kohli Data:
year=list(range(2008,2027))
Virat=[31.8,54.16,47.38,47.62,68.4,52.83,58.55,36.64,92.37,76.84,133.55,59.86,47.88,43.00,27.45,72.47,19.33,65.1,80]
plt.plot(year,Virat)
plt.xlabel("Year")
plt.ylabel("Average Runs")
plt.title("Virat Kohli's Avg. runs per year")
```

```python
year=list(range(2008,2027))
Virat=[31.8,54.16,47.38,47.62,68.4,52.83,58.55,36.64,92.37,76.84,133.55,59.86,47.88,43.00,27.45,72.47,19.33,65.1,80]
Rohit=[25.33,25.50,38.76,55.54,12.92,52.00,52.54,50.93,62.66,71.83,73.57,57.30,57.00,30.00,41.50,52.29,52.33,50.00,20.33]
plt.plot(year,Virat,label="Virat Kohli")
plt.plot(year,Rohit,label="Rohit Sharma")
plt.xlabel("Year")
plt.ylabel("Average Runs")
plt.title("Average runs per year")
plt.legend()
plt.show()
#Can also use more than 2 comaparisons 
```

```python
plt.plot(year,Virat,"ro-",label="Virat Kohli") #'[color][marker][linestyle]'
plt.plot(year,Rohit,"g^:",label="Rohit Sharma")
plt.xlabel("Year")
plt.ylabel("Average Runs")
plt.title("Average runs per year")
plt.legend()
plt.show()
```

```python
plt.plot(year,Virat,color="orange",linestyle="-") #Another way of setting color and linestyle
plt.plot(year,Rohit,color="green",linestyle="-.",linewidth=3) #linewidth can also be set
plt.xlabel("Year")
plt.ylabel("Average Runs")
plt.title("Average runs per year")
plt.tight_layout()# difference show shown better in .py file
plt.grid(True)# grids will be shown
plt.legend(["Virat Kohli","Rohit Sharma"]) #you can also label in legend instead of plot 
plt.show()
```
