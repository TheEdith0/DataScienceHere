```python
#Bins are like range 

import matplotlib.pyplot as plt
ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]
#if bins value not given then it will choose any bins
plt.hist(ages,bins=10,edgecolor="grey")
plt.xlabel("Age Group")
plt.ylabel("Frequency")
plt.title("Frequency of Different age groups of watching Tv")
plt.tight_layout()
plt.show()
```

```python
#bins value in list
plt.hist(ages,bins=[0,10,15,25,30,35,40,50],edgecolor="black",alpha=0.7)

plt.axvline(29,linewidth=2,linestyle="-.",alpha=0.4,color="red")
#to add a vertical line

plt.xlabel("Age Group")
plt.ylabel("Frequency")
plt.title("Frequency of Different age groups of watching Tv")
#grid new functionality:
plt.grid(True,linestyle="--",alpha=0.4,color="black")
plt.show()
```
