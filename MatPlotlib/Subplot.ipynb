```python
import matplotlib.pyplot as plt

x=range(1,30)
y1=[y*2 for y in x]
y2=[y*y for y in x]
y3=[y*y*y for y in x]
y4=[y*8 for y in x]

plt.figure(figsize=(8,6))#Make it bigger

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2X")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Y^2")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Y^3")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("8X")

plt.tight_layout()#acoid collapsing
plt.show()
```

```python
fig, axs =plt.subplots(1,2,figsize=(10,5))

axs[0].plot(x,y1)
axs[0].set_xlabel("X")
axs[0].set_ylabel("Y")
axs[0].set_title("2X")

axs[1].plot(x,y2)
axs[1].set_xlabel("X")
axs[1].set_ylabel("Y")
axs[1].set_title("X^2")

fig.suptitle("Comparison")#super title
fig.tight_layout()
fig.subplots_adjust(top=0.8)# adjust the positioning of subplots
fig.savefig("myplot.png")#saves 

plt.show()
```
