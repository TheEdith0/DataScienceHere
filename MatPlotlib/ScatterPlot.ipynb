```python
import matplotlib.pyplot as plt

study_hours=[3,4,5,6,7,8]
marks=[20,60,30,80,70,90]
colors=["red" if x<50 else "green" for x in marks]
sizes=[x for x in marks]
plt.scatter(study_hours,marks,c=colors,s=sizes)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
c_score=[10,30,40,10,60,90]
plt.scatter(study_hours,marks,c=c_score,cmap="viridis",)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(True)
plt.title("Social score and marks as per study hours")
plt.colorbar(label="Social Score")

for i in range(len(study_hours)):
    plt.annotate(f"S{i+1}",(study_hours[i]+0.1,marks[i]-1))
plt.show()
```

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label='Class A', color='blue')
plt.scatter(class_b_hours, class_b_scores, label='Class B', color='orange')

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
