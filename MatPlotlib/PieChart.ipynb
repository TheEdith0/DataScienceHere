```python
#Pie chart is used in better way when catorgries are less
import matplotlib.pyplot as plt 
plt.style.use("ggplot")
players=["Virat","Jadeja","Rohit","Dhoni"]
runs=[9999,5011,8130,6082]
plt.pie(runs,labels=players)
plt.show()
```

```python
plt.pie(runs,labels=players,colors=["#fa4437","#faac37","#acfa37","#fa37af"],autopct="%1.f%%",shadow=True,explode=[0.1,0,0,0])
plt.show()
```

```python
plt.pie(runs,labels=players,
        colors=["#fa4437","#faac37","#acfa37","#fa37af"],
        autopct="%1.1f%%",
        shadow=True,
        explode=[0.1,0,0,0],
        wedgeprops={"edgecolor":"black","linewidth":1,"linestyle":"--","fill":True}) #"Fill":False will not fill the color in 
plt.show()
```

```python
plt.pie(runs,labels=players,
        colors=["#fa4437","#faac37","#acfa37","#fa37af"],
        autopct="%1.1f%%",
        shadow=True,
        explode=[0.1,0,0,0],
        startangle=30)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
