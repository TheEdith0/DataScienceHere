```python
import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]
 
labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']

plt.figure(figsize=(10,8))
plt.stackplot(days,studying,playing,watching_tv,sleeping,labels=labels,colors=colors,alpha=0.7)
plt.legend(loc="upper left")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Hours spent per day")
plt.grid(True)
plt.show()
#alpha value lies from 0 to 1 : it is transparency
```
