```python
with open("IntialData.txt",encoding="utf8") as f:
    data=f.read()
```

```python
data
```

```python
chunks=data.split("\n\n")
```

```python
chunks[0]
```

```python
def parseData(chunk):
    chunk=chunk.strip()
    sep_chunk=chunk.split('\n')
    followers=float((sep_chunk[2].rstrip(" followers")).replace(",","").replace("M","").replace("K",""))
    if ("K" in sep_chunk[2]):
        followers*=1000
    if ("M" in sep_chunk[2]):
        followers*=1000000
    return {"Username":sep_chunk[0],"Posts":int((sep_chunk[1].rstrip(" posts")).replace(",","")),"Followers":int(followers),"Following":int((sep_chunk[3].rstrip(" following")).replace(",","")),"Name":sep_chunk[4],"Bio":"\n".join(sep_chunk[6:])}
```

```python
all_chunks = [parseData(chunk) for chunk in chunks]
print(all_chunks)
```

```python
import json 
with open("parsed.json","w") as f:
    parsedData=json.dump(all_chunks,f,indent=4)
```
