```python
import json
with open("parsed.json","r") as f:
    datalist=json.load(f)
```

```python
maxposts = 0
person = None

for chunk in datalist:
    if chunk["Posts"] > maxposts:
        maxposts = chunk["Posts"]
        person = chunk

print(person)
    
```
